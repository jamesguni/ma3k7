```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A1 = [[1]]
A2 = [[4, 3],
      [1, 2]]
A3 = [[5, 4, 3],
      [6, 1, 2],
      [7, 8, 9]]
A4 = [[16, 15, 14, 13], 
      [5, 4, 3, 12], 
      [6, 1, 2, 11],
      [7, 8, 9, 10]]
A5=[[17, 16, 15, 14, 13], 
  [18, 5, 4, 3, 12], 
   [19, 6, 1, 2, 11], 
   [20, 7, 8, 9, 10], 
   [21, 22, 23, 24, 25]]
A6 = [[36, 35, 34, 33, 32, 31], 
      [17, 16, 15, 14, 13, 30], 
      [18, 5, 4, 3, 12, 29],
      [19, 6, 1, 2, 11, 28], 
      [20, 7, 8, 9, 10, 27], 
      [21, 22, 23, 24, 25, 26]]
A7 = [[37, 36, 35, 34, 33, 32, 31], 
      [38, 17, 16, 15, 14, 13, 30], 
      [39, 18, 5, 4, 3, 12, 29],
      [40, 19, 6, 1, 2, 11, 28], 
      [41, 20, 7, 8, 9, 10, 27], 
      [42, 21, 22, 23, 24, 25, 26],
     [43, 44, 45, 46, 47, 48, 49]]
det1 = np.linalg.det(A1)
det2 = np.linalg.det(A2)
det3 = np.linalg.det(A3)
det4 = np.linalg.det(A4)
det5 = np.linalg.det(A5)

det6 = np.linalg.det(A6)

det7= np.linalg.det(A7)
y=np.array([det1, det2, det3, det4, det5, det6, det7])
print(y)
```

```python
#Plot determinant values on a graph
x=np.arange(1, 8)
y=np.array([det1, det2, det3, det4, det5, det6, det7])
plt.figure(figsize=(10,6))
plt.plot(x, y)
plt.yscale('symlog')
plt.ylim(-10**6, 10**6)
plt.xlabel("n")
plt.ylabel("det($A_{\mathrm{n}}$)")
plt.title("Distribution of determinants of $A_{\mathrm{n}}$ as n varies ")
plt.grid(True)
plt.show()
```
