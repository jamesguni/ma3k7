# Hat Numbers



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def onesim(n):
    hatarr = np.arange(1, n+1)
    rnumgen = np.random.default_rng()
    while len(hatarr) > 1:
        i1 = rnumgen.integers(0, len(hatarr)) #i1 and i2 denote random indices drawn
        i2 = rnumgen.integers(0, len(hatarr))
        n1 = hatarr[i1] #the elements n1 and n2 indexed by i1 and i2 respectively
        n2 = hatarr[i2]
        min_n1n2 = min(n1, n2) 
        max_n1n2 = max(n1, n2)
        diff = max_n1n2 - min_n1n2 #difference between max and min to be added to array
        hatarr = np.delete(hatarr, [i1, i2]) #delete the two original numbers drawn
        hatarr = np.append(hatarr, diff)
    return hatarr[0]
N = 2024
endnum = onesim(N)
print("The remaining piece after the simulation is:", endnum)

```

```python
sims = 10000
x = [onesim(N) for _ in range(sims)]

plt.hist(x, bins=np.arange(1, N+2), align='left', density=True)
plt.title('Distribution of final pieces between 0 and 2024 inclusive')
plt.xlabel('Final pieces')
plt.ylabel('Probability distribution')
plt.show()
```
